```python
# 2. Import libraries
import whisper
import torch
import librosa
from textblob import TextBlob
from transformers import pipeline
import re
```

```python
# 3. Download audio from YouTube
!pip install yt-dlp
```

```python
!yt-dlp -x --audio-format mp3 "https://www.youtube.com/watch?v=4ostqJD3Psc" -o "call.mp3"
```

```python
# 4. Load model (Whisper for speech-to-text)
model = whisper.load_model("small")
```

```python
# 5. Transcribe with timestamps
result = model.transcribe("call.mp3", verbose=True)
```

```python
# Result will have 'text' + 'segments' (with start, end, text)
print(result.keys())
```

```python
print(result["text"][:500])
```

# ==============================================
# 📊 Analysis
# ==============================================

```python
# Helper vars
segments = result['segments']
```

```python
# 1. Talk-time ratio (based on timestamps)
total_time = segments[-1]['end']
```

```python
speaker_times = {"Speaker1": 0, "Speaker2": 0}
```

```python
# 2️. Count number of questions
questions = [seg for seg in segments if "?" in seg['text']]
```

```python
num_questions = len(questions)
```

```python
# 3️. Longest monologue
longest = max([seg['end'] - seg['start'] for seg in segments])
```

```python
print("Longest monologue duration (sec):", longest)
```

```python
# 4️. Sentiment (overall)
sentiment_analyzer = pipeline("sentiment-analysis")
```

```python
sentiment = sentiment_analyzer(result["text"][:512])[0]
```

```python
print("Sentiment:", sentiment)
```

```python
# 5️. Actionable insight
if num_questions < 3:
    insight = "Sales rep should ask more questions to engage customer."
else:
    insight = "Good number of questions asked, focus on listening more."
```

```python
print("Actionable Insight:", insight)
```
